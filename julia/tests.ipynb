```julia
using WAV
using Plots
using FFTW
using FastTransforms
```

```julia
pwd()
```

```julia
wave, fs = wavread(raw"..\samples\basses\bass01.wav")
fs = convert(Int64, fs)
```

```julia
println(size(wave))
println(fs)
```

```julia
duration = size(wave)[1] / fs
println("duration: $(duration)s")
```

```julia
plot(wave)
```

```julia
f = fft(wave)
freqs = fftfreq(length(wave), fs)
```

```julia
plot(freqs, abs.(f), xlims=(0, 4e3))
```
